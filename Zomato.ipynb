```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

Observations
1.) RAtings- 4.5 to 4.9 -> Excellent
2.) Ratings- 4.0 to 4.4 -> Very Good
3.) Ratings- 3.5 to 3.9 -> Good
4.) Ratings- 2.5 to 3.4 -> Average
5.) Ratings- 2.0 to 2.4 -> Poor



```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings,hue='Rating color',palette=['White','Red','Orange','Yellow','Green','Green'])
```

Observation 
1.) Not rated count is very high.
2.) Maximum number of ratings are between 2.5 to 3.4.


```python
sns.countplot(x='Rating color',data=ratings,palette=['White','Red','Orange','Yellow','Green','Green'])
```
